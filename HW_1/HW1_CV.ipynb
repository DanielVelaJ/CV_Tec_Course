# Capture
1. Capture an Image 
Use escape to exit and space to capture an image


```python
cam = cv2.VideoCapture(0)
cv2.namedWindow('capture')
img_counter=0
while True:
    ret,frame=cam.read()
    if not ret:
        print('failed to capture frame')
        break
    cv2.imshow('capture',frame)
    k = cv2.waitKey(1)
    if k == 27:
        print('Escape hit...closing')
        break
    elif k == 32:
        #space pressed
        img_name='capture.png'
        cv2.imwrite(img_name,frame)
        print(f'image captured and saved')
cam.release()# clos everything
cv2.destroyAllWindows()
    
```

2. Convert to Gray Scale


```python
image = cv2.imread('capture.png')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
fig,axes=plt.subplots(1,2,figsize=(15,5))
axes[0].imshow(image[:,:,[2,1,0]]) #image channels are reordered
plt.title('Original')
map=axes[1].imshow(gray,cmap='binary')
plt.title('Gray')
fig.colorbar(map)
```


3. Invert the Gray Scale (Negative Image)

```python
inverted=(gray)*(-1)+255 # Invert the colormap
plt.imshow(inverted,cmap='Greys')
plt.colorbar()
```

5. Using a line pattern
        

```python
import cv2 as cv
cap = cv.VideoCapture(0)

x = 0

if not cap.isOpened():
    print("Cannot open camera")
    exit()


while True:
    x = x+1
    # Capture frame-by-frame
    ret, frame = cap.read()

    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Our operations on the frame come here

    cv.imshow('ColorF', frame)
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    gray.astype(float)
    # Display the resulting frame
    cv.imshow('Gray', gray)

    b, g, r = cv.split(frame)
    cv.imshow('BlueF', b)
    cv.imshow('GreenF', g)
    cv.imshow('RedF', r)

    si, sj, ch = frame.shape
    dim = (int(sj/2), int(si/2))
    frame = cv.resize(frame, dim, interpolation=cv.INTER_NEAREST)

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    gray.astype(float)
    # Display the resulting frame
    cv.imshow('Gray/2', gray)

    if cv.waitKey(1) == ord('q'):
        break
# When everything done, release the capture
cap.release()
cv.destroyAllWindows()

cap = cv.VideoCapture(0)

x = 0

if not cap.isOpened():
    print("Cannot open camera")
    exit()


while True:
    x = x+1
    # Capture frame-by-frame
    ret, frame = cap.read()

    # if frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break
    # Our operations on the frame come here

    cv.imshow('ColorF', frame)
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    gray.astype(float)
    # Display the resulting frame
    cv.imshow('Gray', gray)

    b, g, r = cv.split(frame)
    cv.imshow('BlueF', b)
    cv.imshow('GreenF', g)
    cv.imshow('RedF', r)

    si, sj, ch = frame.shape
    dim = (int(sj/2), int(si/2))
    frame = cv.resize(frame, dim, interpolation=cv.INTER_NEAREST)

    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
    gray.astype(float)
    # Display the resulting frame
    cv.imshow('Gray/2', gray)

    if cv.waitKey(1) == ord('q'):
        break
# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
```
